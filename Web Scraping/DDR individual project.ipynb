(1)Can you trigger the sorting change directly by modifying only the URL in your browser’s address bar?  If so, how?

Yes I can trigger the sorting change directly by modifying the keyword behind "sort" in the URL. (2)Explain what type of request is made when you change the sort order (GET or POST).

In our case, we are using GET request when we change the sort order in our URL. GET request is used to retrieve data from a resource and POST request is used to send data to a server in order to update a resource. Hence, when we change the sorting order, the browser sends a GET request to the server with the updated URL. The server then responds with a new page where the listings are sorted by oldest date.

(3)Extract the first 250 unique listing URLs and save them to a list.  Consider the pagination feature of Craigslist to navigate through pages.  Explain your strategy.

In order to grab the 250 listing URLs, I generated a while loop to check it. Using a while loop to keep running until we have 250 links. After extracting all linkes from the page, use if to check if we have enough linkes. If we don't have enough linkes, increment the page number and continue extract linkes from next page until it has 250 URLs.

```python
from bs4 import BeautifulSoup
import requests
import time 

#method 1
#create a empty list to store all the links
links = []
page_number = 0
while len(links) < 250:
    url = f'https://sfbay.craigslist.org/search/zip?sort=date#search=1~gallery~{page_number}~0'
    headers = {'user-agent':'Mozilla/5.0'}
    page = requests.get(url, headers = headers)
    soup = BeautifulSoup(page.content, 'html.parser')
    result = soup.select('li', class_='cl-search-resultcl-search-view-mode-gallery')
    #using for loop to find all links
    for item in result:
        a_tag = item.select_one('a[href]')
        if a_tag:
            link = a_tag['href']
            #print(link)
            if link not in links:
                links.append(link)
        if len(links) >= 250:
                break
    if len(links) >= 250:  # Stop if we have collected 250 links
        break
    page_number += 1
    time.sleep(10)
                
print(links)
print(len(links))
```

```python
for filename in os.listdir(directory):
    if filename.endswith(".html"):

        # Construct the full file path

        filepath = os.path.join(directory, filename)

        # Read file to string

        with open(filepath, 'r', encoding='utf-8') as file:

            html = file.read()
            
        #scrap all the content in file
            soup = BeautifulSoup(html, 'html.parser')
            
        # look for the title for each file
            title = soup.find('title').text
        
        # look for image link
            img_link = soup.find('img')
            img_link = img_link['src'] if img_link and 'src' in img_link.attrs else "No image found"
            
        # look for description
            description_tag = soup.find('section', {'id': 'postingbody'})
            description = ' '.join(description_tag.text.strip().splitlines()) if description_tag else "No description found"

        # look for post id
            post_id = soup.select('p.postinginfo')
            if len(post_id) > 0:
                post_id = post_id[1].text
            else: 
                post_id = 'post_id : No post id'
            
                
        #look for post date
            post_date = soup.select('p.postinginfo.reveal')
            if len(post_date) > 0:
                post_date = post_date[1].text
            else: 
                post_id = 'post_date : No post time'
             
        
        #look for update date
            update_date = soup.select('p.postinginfo')
            
            if len(update_date) > 2:
                update_date_text = update_date[3].text.strip()
                if update_date_text.startswith('updated'):
                    update_date = update_date_text
                else:
                    update_date = "No updated date found"
            else:
                update_date = "No updated date found"

            print(f'title: {title}\nimage link:{img_link}\ndescription:{description}\n{post_id}\n{post_date}\n{update_date}\n ')
```

#### (2) Exploring the Login Mechanism

In the 'form' tag in the login page, it includes three main fields, username input field, password input field, and submit button. #### (3）Analyzing Network Traffic for Login Request

According to the 'Network', login use POST method instead of GET. Both POST and GET methods are request methods. However, POST is ususlly used when we need to submit our actions to server while GET is usually used in simple retrieve data. In our situation, since we need to submit our username and password to server, we need to use POST. In addition, if we use GET method, the data will be shown in URL, which means everyone can know my username and password by looking at the my broswer's address bar. My username and password is a privacy data, so we don't want it to be shown in the URL. Using POST method is a good way to solve this problem. POST helps maintain privacy for personal information.

The structure of payload is similar with form field, including user[login], user[password], and so on. Different with the form field, we can directly see our username and password in payload page. And the value, username and password, is correspond to the name attributes of the input fields.
